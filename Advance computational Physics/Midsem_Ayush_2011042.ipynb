# Problem 1

```python
f1 = lambda x : log(x/2) - sin(5*x/2) # the function given
df1 = lambda x : 1/(x) - (5*cos(5*x/2))/2 # the derivative of the function
init_interval = [1.5,2.5] # the starting interval guess to find the root
e = 6 # i.e the value of accuracy is upto 10^-6

# checking whether the initial interval is yielding any root
Regula_falsi(f1,init_interval,e)
```

```python
# Since the initial interval is not correct and won't yield any root, 
# applying Bracketing function to get the right interval

new_interval = Bracketing(f1,init_interval)
print(f'The numerically obtained new interval is : {new_interval}')
```

```python
# Trying again with new interval
Regula_falsi(f1,new_interval,e)
```

```python
# Trying with Newton-Raphson method
guess_roots = [1.5,2.5,3.5] # initial guesses for the roots
#Note: Newton-Raphson method is sensitive to the initial guess, so we need to try with different initial guesses
for i in guess_roots:
    print(f"\nTrying with initial guess = {i}")
    Newton_Raphson(f1,df1,i,e)
    
# Newton-Raphson yeilds roots faster compared to Regula-Falsi method 
```

# Problem 4

The centre of mass of the rod can be calculated using:

$$x_c = \int x\lambda(x)dx / \int \lambda(x)dx$$

```python
#so we just have to numerically evaluate two integrals in the boundary [0,2]
ximi = lambda x: x**3
mi = lambda x : x**2

#using Simpson method to solve the required integrals and finding the centre of mass
COM = Simpson_int(ximi,0,2,20)/Simpson_int(mi,0,2,20)

print(f"The position of the center of mass of the given distribution is {COM: 1.4f} m away from the origin")
```
